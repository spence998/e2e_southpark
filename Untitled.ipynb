### Plot and character description functions

```python
def plot_llama(prompt):
    llm = AutoModelForCausalLM.from_pretrained(
        "C:\dev\models\llama2\llama-2-13b-chat.Q5_K_M.gguf",
        model_type="llama",
        gpu_layers=50,
    )
    
    return llm(
        "[INST] <<SYS>>You are a story writer creating one scene in the world of South Park. "
        "The scene will consist of two characters and will be simple and mostly involve "
        "simple dialogue about the topic involving each character speaking twice. "
        "Given an idea, you will generate a simple plotline in under two sentences and 50 words.. "
        f"<</SYS>>The plot is: {prompt}[/INST]",
    )

def character_extraction_llama(prompt):
    llm = AutoModelForCausalLM.from_pretrained(
        "C:\dev\models\llama2\llama-2-13b-chat.Q5_K_M.gguf",
        model_type="llama",
        gpu_layers=50
    )
    
    return llm(
        "[INST] <<SYS>>You bot who can analyse a South Park plot and can respond with only two names. "
        "You are given a scene and you are to find out who are the two characters in the scene. "
        "You will write two words, the names of each character split by a comma and nothing else.<</SYS>>"
        "The scene is: <<Kenny and Cartman are playing football when Kenny falls over in a puddle.>> "
        "The two characters in the scene are: Kenny, Cartman"
        f"The scene is: <<{prompt}>>. The two characters in the scene are: [/INST]",
    )

def character_description_llama(character, plot):
    llm = AutoModelForCausalLM.from_pretrained(
        "C:\dev\models\llama2\llama-2-13b-chat.Q5_K_M.gguf",
        model_type="llama",
        gpu_layers=50
    )
    
    return llm(
        "[INST] <<SYS>>You are given a scene and one character in the scene. You will describe what the character looks like in 8 words."
        "Only include anything particularly noteworthy. Again only 8 words."
        f"<</SYS>>The character is: {character}, the scene is {plot}[/INST]",
    )

def character_items_llama(character, plot):
    llm = AutoModelForCausalLM.from_pretrained(
        "C:\dev\models\llama2\llama-2-13b-chat.Q5_K_M.gguf",
        model_type="llama",
        gpu_layers=50
    )
    
    return llm(
        "[INST] <<SYS>>You are given a scene and one character in the scene. "
        "You will write one word for anything they are holding."
        f"<</SYS>>The character is: {character}, the scene is {plot}[/INST]",
    )

def background_description_llama(plot):
    llm = AutoModelForCausalLM.from_pretrained(
        "C:\dev\models\llama2\llama-2-13b-chat.Q5_K_M.gguf",
        model_type="llama",
        gpu_layers=50,
    )
    
    return llm(
        "[INST] <<SYS>>You are a story writer creating one scene in the world of South Park. "
        "Given a plot, you will write a 10 word sentence describing the most significant background "
        "objects. This may include buildings objects, sofas, cars or anything else significant. "
        "Focus on fewer items but be more descriptive on them. You will not include any characters in your description."
        f"<</SYS>>The scene is {plot}[/INST]",
    )

def get_dialogue(plot):
    plot_just_speech = re.sub("[\(\[].*?[\)\]]", "", plot)[1:]
    speech = re.split("[A-Za-z]+:+ '", plot_just_speech)[1:]
    who_spoke = re.findall("[A-Za-z]+:+", plot_just_speech)
    who_spoke = [w[:-1] for w in who_spoke]
    return who_spoke, speech

def get_non_dialogue(plot):
    plot_just_speech = ", ".join(re.findall("\((.*?)\)", plot))
    return plot_just_speech
```

### South park style image generation functions

```python
def create_image(prompt, image_name):
    pipe = StableDiffusionPipeline.from_pretrained(
        r"C:\dev\models\SouthParkStyle",
        torch_dtype=torch.float16,
    )
    pipe = pipe.to("cuda")
       
    while True:
        images = pipe(
            prompt=prompt,
            num_images_per_prompt=1,
        )
        if not any(images.nsfw_content_detected):
            if not os.path.isdir('images'):
                os.mkdir("images")
            for i, img in enumerate(images.images):
                img.save(f"images/{image_name}.png")
            return f"images/{image_name}.png"
```

### Editting the image based on a prompt functions

```python
def get_model_pipeline(model_path):
    pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
        model_path,
        torch_dtype=torch.float16,
        safety_checker=None,
    )
    pipe.to("cuda")
    pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(
        pipe.scheduler.config,
    )
    return pipe


def load_image(path):
    image = PIL.Image.open(path)
    image = PIL.ImageOps.exif_transpose(image)
    image = image.convert("RGB")
    return image


def open_mouth(path, save_path):
    model_path = r"C:\dev\models\instruct-pix2pix"
    pipe = get_model_pipeline(model_path)

    image = load_image(path)
    
    prompt = f"Give south park style open mouth"

    images = pipe(
        prompt,
        image=image,
        num_inference_steps=10,
        num_images_per_prompt=1,
        image_guidance_scale=2.5,
        guidance_scale=7.5,
    ).images

    for i, img in enumerate(images):
        img.save(save_path)


def give_item(path, save_path, item):
    model_path = r"C:\dev\models\instruct-pix2pix"
    pipe = get_model_pipeline(model_path)

    image = load_image(path)
    
    prompt = f"In south park style, give the character a {item}"

    images = pipe(
        prompt,
        image=image,
        num_inference_steps=10,
        num_images_per_prompt=1,
        image_guidance_scale=2.5,
        guidance_scale=7.5,
    ).images

    for i, img in enumerate(images):
        img.save(save_path)
```

### Removing backgrounds of characters

```python
def remove_background(path):
    input = Image.open(path)
    output = remove(input)
    output.save(path)
```

```python
reference_cartman = [r"voices\cartman.wav"]
reference_butters = [r"voices\butters.wav"]

def generate_speech(reference_clips, text, output_file):
    reference_clips = [utils.audio.load_audio(p, 22050) for p in reference_clips]
    tts = api.TextToSpeech()
    pcm_audio = tts.tts_with_preset(
        text,
        voice_samples=reference_clips,
        preset="fast",
    )
    
    torchaudio.save(
        output_file,
        pcm_audio.squeeze(0).cpu(),
        24000,
    )
```
